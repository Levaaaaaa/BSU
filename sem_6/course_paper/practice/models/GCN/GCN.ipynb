```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import dgl
from dgllife.model import GCNPredictor
from dgllife.utils import smiles_to_bigraph
from rdkit import Chem
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from torch.utils.data import DataLoader
from tqdm import tqdm
import numpy as np
```

```python

# Загрузка ESOL
url = "https://deepchemdata.s3-us-west-1.amazonaws.com/datasets/delaney-processed.csv"
df = pd.read_csv(url)
df = df[df['smiles'].apply(lambda s: Chem.MolFromSmiles(s) is not None)]
```

```python
# Разделение
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# Функция для извлечения признаков
def extract_node_features(mol):
    atom_feats = []
    for atom in mol.GetAtoms():
        atom_feats.append(atom.GetAtomicNum())
    return torch.tensor(atom_feats).unsqueeze(-1).float()
```

```python
# Создание датасета
def create_dataset(df):
    graphs = []
    feats = []
    labels = []
    for _, row in df.iterrows():
        mol = Chem.MolFromSmiles(row['smiles'])
        g = smiles_to_bigraph(row['smiles'])
        f = extract_node_features(mol)
        graphs.append(g)
        feats.append(f)
        labels.append([row['measured log solubility in mols per litre']])
    return graphs, feats, torch.tensor(labels, dtype=torch.float32)
```

```python
def collate(samples):
    graphs, feats, labels = map(list, zip(*samples))
    bg = dgl.batch(graphs)
    feats = torch.cat(feats)
    labels = torch.stack(labels).squeeze()
    return bg, feats, labels
```

```python
train_graphs, train_feats, train_labels = create_dataset(train_df)
test_graphs, test_feats, test_labels = create_dataset(test_df)
train_graphs = [dgl.add_self_loop(g) for g in train_graphs]
test_graphs = [dgl.add_self_loop(g) for g in test_graphs]
```

```python
# Обёртка датасета
class MoleculeDataset(torch.utils.data.Dataset):
    def __init__(self, graphs, feats, labels):
        self.graphs = graphs
        self.feats = feats
        self.labels = labels
    def __len__(self):
        return len(self.graphs)
    def __getitem__(self, idx):
        return self.graphs[idx], self.feats[idx], self.labels[idx]
```

```python
train_loader = DataLoader(MoleculeDataset(train_graphs, train_feats, train_labels),
                          batch_size=32, shuffle=True, collate_fn=collate)
test_loader = list(zip(test_graphs, test_feats, test_labels))
```

```python
# Объявление модели
model = GCNPredictor(in_feats=1, hidden_feats=[64, 64], activation=[F.relu, F.relu])
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.MSELoss()
```

```python
# Обучение
model.train()
for epoch in range(50):
    model.train()
    total_loss = 0
    for bg, feats, labels in train_loader:
        preds = model(bg, feats)
        loss = loss_fn(preds.view(-1), labels.view(-1))
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}: Loss {total_loss / len(train_loader):.4f}")
```

```python
# Тестирование
model.eval()
preds = []
y_true = []
with torch.no_grad():
    for g, f, y in test_loader:
        pred = model(g, f)
        preds.append(pred.item())
        y_true.append(y.item())
```

```python
rmse = np.sqrt(mean_squared_error(y_true, preds))
r2 = r2_score(y_true, preds)
print(f"Test RMSE: {rmse:.4f}, R²: {r2:.4f}")
```

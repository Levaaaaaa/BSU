class BatchNormalization(Module):
    EPS = 1e-3

    def __init__(self, alpha=0.):
        super(BatchNormalization, self).__init__()
        self.alpha = alpha
        self.moving_mean = 0.
        self.moving_variance = 1.
        self.training = True
        self.gamma = 1
        self.beta = 0

    def _compute_output(self, input):
        # Your code goes here. ################################################
        # output = ...
        if self.training:
            batch_mean = np.mean(input, axis=0)
            batch_variance = np.var(input, axis=0)

            output = (input.T - batch_mean) / np.sqrt(batch_var + this.EPS)
            
            self.moving_mean = self.moving_mean * alpha + batch_mean * (1 - alpha)
            self.moving_variance = self.moving_variance * alpha + batch_variance * (1 - alpha)

            return self.gamma * output + self.beta, (output, batch_mean, batch_variance)
        else :
            output = (input.T - self.moving_mean) / np.sqrt(self.moving_variance + this.EPS)
            return self.gamma * output + self.beta

    def _compute_input_grad(self, input, output_grad):
        # Your code goes here. ################################################
        x_normalized, batch_mean, batch_var = self._compute_output(input)
        N, D = output_grad.shape

        dgamma = np.sum(output_grad * x_normalized, axis=0)
        dbeta = np.sum(output_grad, axis=0)

        dx_normalized = output_grad * self.gamma

        dvar = np.sum(dx_normalized * (input - batch_mean) * -0.5 * np.power(batch_var + self.EPS, -1.5), axis=0)
        dmean = np.sum(dx_normalized * -1 / np.sqrt(batch_var + self.EPS), axis=0) + dvar * np.mean(-2 * (input - batch_mean), axis=0)

        dx = dx_normalized / np.sqrt(batch_var + self.EPS) + (dvar * 2 * (input - batch_mean) / N) + (dmean / N)

        grad_input = dx, dgamma, dbeta
        return grad_input

    def __repr__(self):
        return "BatchNormalization"

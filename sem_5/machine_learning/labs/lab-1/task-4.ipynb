```python
import numpy as np
import pandas as pd

base = './'
data = pd.read_csv(base + 'organisations.csv')
rubrics = pd.read_csv(base + 'rubrics.csv')
rubrics_dict = dict(zip(rubrics['rubric_name'], rubrics['rubric_id']))
cafe_id = rubrics_dict['Кафе']
new_data = data.dropna(subset=['average_bill'])
new_data = new_data.loc[new_data['average_bill'] <= 2500]
new_data = new_data.loc[new_data['rubrics_id'].str.contains(str(cafe_id))]
av = new_data[['city', 'average_bill']].groupby(by=['city']).mean()
print(av)
av.loc['msk'] - av.loc['spb']
```
